```python
#案例一 爬取百度文库的内容：打开->获取内容->关闭

from selenium.webdriver import Chrome
#from selenium.webdriver.common.keys import Keys#输入关键词
from selenium import webdriver
import time
import random


#1.创建浏览器对象
browser = Chrome()
#2.打开目标网址
browser.get('http://wenku.baidu.com/view/ff9655cdcdc789eb172ded630b1c59eef9c79a2f.html')
#3.加入时间延迟，给浏览器缓冲的时间
time.sleep(random.uniform(3,4))
#3.找到需要的元素
content_p1 = browser.find_element_by_xpath('//*[@id="pageNo-1"]').text
content_p2 = browser.find_element_by_xpath('//*[@id="pageNo-2"]').text

print(content_p1+content_p2)

# 退出浏览器
browser.quit()
```

```python
#案例二 爬取京东商城天梭的价格：打开->搜索->获取内容->关闭
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
import random
import csv


#1.创建浏览器对象
browser = Chrome()
#2.打开一个网址
browser.get('http://jd.com')
#找到需要的元素，点击它，方法为find_element_by_xpath
browser.find_element_by_xpath('//*[@id="key"]').send_keys('天梭',Keys.ENTER)

#让程序等待浏览器加载的时间设置
time.sleep(random.uniform(1,2))

#将页面拖至最下方,将待加载的内容显全
js = 'var action = document.documentElement.scrollTop=10000'
#执行js操作
browser.execute_script(js)

time.sleep(random.uniform(1,2))

#获取多个节点，不可以用element，而是elements
li_goods = browser.find_elements_by_xpath('//*[@id="J_goodsList"]/ul/li')




for li in li_goods:
    
    time.sleep(random.uniform(1, 2))
    
    li_price = li.find_element_by_xpath('./div/div[2]/strong/i').text
    li_name = li.find_element_by_xpath('./div/div[3]/a/em').text
    li_comm = li.find_element_by_xpath('./div/div[4]/strong').text
    print([li_price,li_name,li_comm])
    '''
    with open (r'./text.txt','a',encoding = 'utf-8') as result:
        result.write(li_price+li_name+li_comm+'\n')  
       

    with open (r'./csvres.csv','a+',newline = '') as fw:
        fw_csv = csv.writer(fw)
        fw_csv.writerow([li_price,li_name,li_comm])
    ''' 


browser.quit()
```

```python
#案例3 拉勾网的页面交互操作
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
import random

#1.创建Chrome 对象
browser = Chrome()
#2.发起网页请求
browser.get('http://lagou.com')

#3.对浏览器发出指令
browser.find_element_by_xpath('//*[@id="changeCityBox"]/ul/li[1]/a').click()

#加载随机延迟时间
time.sleep(random.uniform(1, 2))

browser.find_element_by_xpath('//*[@id="search_input"]').send_keys('python',Keys.ENTER)
#加载随机延迟时间
time.sleep(random.uniform(1, 2))

#点击子页面,未更迭前的代码//*[@id="s_position_list"]/ul/li[1]/div[1]/div[1]/div[1]/a/h3
browser.find_element_by_xpath('//*[@id="jobList"]/div[1]/div[1]/div[1]/div[1]/div[1]/a').click()
#加载随机延迟时间
time.sleep(random.uniform(1, 2))
#switch to方法进入子页面,若否selenium自己不进入
browser.switch_to_window(browser.window_handles[-1])
#加载随机延迟时间
time.sleep(random.uniform(1, 2))

job_detail = browser.find_element_by_xpath('//*[@id="job_detail"]/dd[2]/div').text

#print(job_detail)

browser.close()#关掉已经爬取数据的子窗口

browser.switch_to_window(browser.window_handles[0])#关掉子窗口后视角变更为原窗口中

browser.quit()
```

```python
#案例四更简单的内容理解找到所有节点和找到某个节点

from selenium import webdriver
from selenium.webdriver import Chrome
import time
import random

#1.创建Chrome 对象
browser = Chrome()
#2.发起网页请求
browser.get('http://rsc.nwu.edu.cn/tzgg.htm')
#加载随机延迟时间
time.sleep(random.uniform(1,2))
#找到多个元素elements，xpath的路径末尾写道内容开始标签
notic_lst = browser.find_elements_by_xpath('/html/body/div[1]/div[3]/section/div[3]/ul/li')
                                            
# ./代表全部内容
for i in notic_lst:
    title = i.find_element_by_xpath('./div[1]/a/h3/b').text
    print(title)
    
browser.quit()
```
